```python
import pandas as pd
import random
import collections
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import math
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#first data
df = pd.read_csv('data/events_countries.txt',sep='\t' )

#second data
dfFull = pd.read_csv('data/events_countries_sources_years.txt',sep='\t' )
dfFull.dropna(inplace=True)
dropYearList=[2003,2004,2005,2006,2020,2021,2022,2025,2026,2029]
dfFull=dfFull[~dfFull['year'].isin(dropYearList)]
yearList=[]
for i in range(2007,2020):
    yearList.append(i)
    
#add continet
continentMap={}
for _, r in df.iterrows():
    if r['un_country'] not in continentMap.keys():
        continentMap[r['un_country']]=r['macro_region']
macroList=[]
for _, r in dfFull.iterrows(): 
    k=r['un_country']
    macroList.append(continentMap[k])
dfFull['macro_region']=macroList


#population data add continent

countryName={
    "bahamas":"the bahamas",
    "congo, dem. rep.":"democratic republic of the congo",
    "cote d'ivoire":"côte d'ivoire",
    "czech republic":"the czech republic",
    "kyrgyz republic":"kyrgyzstan",
    "lao":"laos",
    "micronesia, fed. sts.":"federated states of micronesia",
    "myanmar":"myanmar (burma)",
    "sao tome and principe":"são tomé and príncipe",
    "slovak republic":"slovakia",
    "st. kitts and nevis":"saint kitts and nevis",
    "st. lucia":"saint lucia",
}
unfoundCountry={
    "central african republic":"Africa",
    "chad":"Africa",
    "congo, rep.":"Africa",
    "comoros":"Africa",
    'djibouti':"Africa",
    'dominica':'Americas',
    'equatorial guinea':"Africa",
    'gambia':"Africa",
    'grenada':"Americas",
    'guinea-bissau':"Africa",
    'kiribati':"Oceania",
    'marshall islands':"Oceania",
    'nauru':"Oceania",
    "niger":"Africa",
    'palau':"Oceania",
    "palestine":"Asia",
    'samoa':"Oceania",
    "st. vincent and the grenadines":"Americas",
    "tonga":"Oceania",
    "tuvalu":"Oceania"
}
dfPop = pd.read_csv('data/population_total.csv',sep=',' )
keepColumns=['country']
for c in yearList:
    keepColumns.append(str(c))
dfPop=dfPop[keepColumns]
dfPop.head(10)

popContinentList=[]
for _, r in dfPop.iterrows(): 
    k=r['country'].lower()
    
    if k in continentMap.keys():
        popContinentList.append(continentMap[k])
    elif k in countryName.keys():
        popContinentList.append(continentMap[countryName[k]])
    else:
        
        popContinentList.append(unfoundCountry[k])
    
dfPop['macro_region']=popContinentList
```

```python
import pandas as pd
import random
import collections
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import math
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#first data
df = pd.read_csv('data/events_countries.txt',sep='\t' )

#second data
dfFull = pd.read_csv('data/events_countries_sources_years.txt',sep='\t' )
dfFull.dropna(inplace=True)
dropYearList=[2003,2004,2005,2006,2020,2021,2022,2025,2026,2029]
dfFull=dfFull[~dfFull['year'].isin(dropYearList)]
yearList=[]
for i in range(2007,2020):
    yearList.append(i)
    
#add continet
continentMap={}
for _, r in df.iterrows():
    if r['un_country'] not in continentMap.keys():
        continentMap[r['un_country']]=r['macro_region']
macroList=[]
for _, r in dfFull.iterrows(): 
    k=r['un_country']
    macroList.append(continentMap[k])
dfFull['macro_region']=macroList


#population data add continent

countryName={
    "bahamas":"the bahamas",
    "congo, dem. rep.":"democratic republic of the congo",
    "cote d'ivoire":"côte d'ivoire",
    "czech republic":"the czech republic",
    "kyrgyz republic":"kyrgyzstan",
    "lao":"laos",
    "micronesia, fed. sts.":"federated states of micronesia",
    "myanmar":"myanmar (burma)",
    "sao tome and principe":"são tomé and príncipe",
    "slovak republic":"slovakia",
    "st. kitts and nevis":"saint kitts and nevis",
    "st. lucia":"saint lucia",
}
unfoundCountry={
    "central african republic":"Africa",
    "chad":"Africa",
    "congo, rep.":"Africa",
    "comoros":"Africa",
    'djibouti':"Africa",
    'dominica':'Americas',
    'equatorial guinea':"Africa",
    'gambia':"Africa",
    'grenada':"Americas",
    'guinea-bissau':"Africa",
    'kiribati':"Oceania",
    'marshall islands':"Oceania",
    'nauru':"Oceania",
    "niger":"Africa",
    'palau':"Oceania",
    "palestine":"Asia",
    'samoa':"Oceania",
    "st. vincent and the grenadines":"Americas",
    "tonga":"Oceania",
    "tuvalu":"Oceania"
}
dfPop = pd.read_csv('data/population_total.csv',sep=',' )
keepColumns=['country']
for c in yearList:
    keepColumns.append(str(c))
dfPop=dfPop[keepColumns]
dfPop.head(10)

popContinentList=[]
for _, r in dfPop.iterrows(): 
    k=r['country'].lower()
    
    if k in continentMap.keys():
        popContinentList.append(continentMap[k])
    elif k in countryName.keys():
        popContinentList.append(continentMap[countryName[k]])
    else:
        
        popContinentList.append(unfoundCountry[k])
    
dfPop['macro_region']=popContinentList

#making the graph, subpolot
dfPopMelt = pd.melt(dfPop,id_vars=['country','macro_region'],var_name='year',value_name='pop')
dfPopMelt.sort_values(by=['country'],inplace=True)

dfPopAisa=dfPopMelt[dfPopMelt['macro_region']=='Asia']
dfPopAisa=dfPopAisa.groupby(['year'],as_index=False)[['pop']].sum()

dfPopAmericas=dfPopMelt[dfPopMelt['macro_region']=='Americas']
dfPopAmericas=dfPopAmericas.groupby(['year'],as_index=False)[['pop']].sum()

dfPopEurope=dfPopMelt[dfPopMelt['macro_region']=='Europe']
dfPopEurope=dfPopEurope.groupby(['year'],as_index=False)[['pop']].sum()

dfPopAfrica=dfPopMelt[dfPopMelt['macro_region']=='Africa']
dfPopAfrica=dfPopAfrica.groupby(['year'],as_index=False)[['pop']].sum()

dfPopWorld=dfPopMelt.groupby(['year'],as_index=False)[['pop']].sum()


# 'Asia', 'Europe', 'Africa', 'Americas'
dfFullAsia=dfFull[dfFull['macro_region']=='Asia']
dfFullAsia=dfFullAsia.groupby(['year'],as_index=False)[['n']].sum()

dfFullAmericas=dfFull[dfFull['macro_region']=='Americas']
dfFullAmericas=dfFullAmericas.groupby(['year'],as_index=False)[['n']].sum()

dfFullEurope=dfFull[dfFull['macro_region']=='Europe']
dfFullEurope=dfFullEurope.groupby(['year'],as_index=False)[['n']].sum()

dfFullAfrica=dfFull[dfFull['macro_region']=='Africa']
dfFullAfrica=dfFullAfrica.groupby(['year'],as_index=False)[['n']].sum()

dfFullWorld=dfFull.groupby(['year'],as_index=False)[['n']].sum()

fig = make_subplots(rows=3, cols=2, start_cell="bottom-left",horizontal_spacing = 0.12,vertical_spacing = 0.13,
                   shared_xaxes=True,
                   specs=[[{"secondary_y": True}, {"secondary_y": True}],
                          [{"secondary_y": True}, {"secondary_y": True}],
                          [{"colspan": 2,"secondary_y": True}, None]],
                    subplot_titles=("Asia","Americas", "Europe","Africa","Worldwide"))
#------------------------------------------------------------------------------
fig.add_trace(go.Scatter(
                x=dfFullAsia['year'],
                y=dfFullAsia['n'],
                name='Asia Events',
                fill='tozeroy',
                hovertemplate ="year: <b>%{x}</b><br>Region: Asia<br>Events:%{x}<br>",
             ),
              row=1, col=1,)
fig.add_trace(go.Scatter(
                x=dfPopAisa['year'],
                y=dfPopAisa['pop'],
                name='Asia Populaiton',
                hovertemplate ="year: <b>%{x}</b><br>Region: Asia<br>Population:%{x}<br>",
             ),
              row=1, col=1,secondary_y=True,)
fig.update_xaxes(
                 tickangle=60,
                 nticks=7,
                 row=1, col=1)
#------------------------------------------------------------------------------
fig.add_trace(go.Scatter(
                x=dfFullAmericas['year'],
                y=dfFullAmericas['n'],
                name='America Events',
                fill='tozeroy',
                hovertemplate ="year: <b>%{x}</b><br>Region: America<br>Events:%{x}<br>",
             ),
              row=1, col=2)

fig.add_trace(go.Scatter(
                x=dfPopAmericas['year'],
                y=dfPopAmericas['pop'],
                name='America Populaiton',
                hovertemplate ="year: <b>%{x}</b><br>Region: America<br>Population:%{x}<br>",
             ),
              row=1, col=2,secondary_y=True,)
fig.update_xaxes(
                 tickangle=60,
                 nticks=7,
                 row=1, col=2)

#------------------------------------------------------------------------------
fig.add_trace(go.Scatter(
                x=dfFullEurope['year'],
                y=dfFullEurope['n'],
                name='Europe Events',
                fill='tozeroy',
                hovertemplate ="year: <b>%{x}</b><br>Region: Europe<br>Events:%{x}<br>",
             ),
              row=2, col=1)
fig.add_trace(go.Scatter(
                x=dfPopEurope['year'],
                y=dfPopEurope['pop'],
                name='Europe Populaiton',
                hovertemplate ="year: <b>%{x}</b><br>Region: Europe<br>Population:%{x}<br>",
             ),
              row=2, col=1,secondary_y=True,)
fig.update_xaxes(
#                  tickvals = yearList,
                 tickangle=60,
                 nticks=7,
                 row=2, col=1)
#------------------------------------------------------------------------------
fig.add_trace(go.Scatter(
                x=dfFullAfrica['year'],
                y=dfFullAfrica['n'],
                name='Africa Events',
                fill='tozeroy',
                hovertemplate ="year: <b>%{x}</b><br>Region: Africa<br>Events:%{x}<br>",
             ),
              row=2, col=2)
fig.add_trace(go.Scatter(
                x=dfPopAfrica['year'],
                y=dfPopAfrica['pop'],
                name='Africa Populaiton',
                hovertemplate ="year: <b>%{x}</b><br>Region: Africa<br>Population:%{x}<br>",
             ),
              row=2, col=2,secondary_y=True,)
fig.update_xaxes(
                 tickangle=60,
                 nticks=7,
                 row=2, col=2)

#------------------------------------------------------------------------------
fig.add_trace(go.Scatter(
                x=dfFullWorld['year'],
                y=dfFullWorld['n'],
                name='Worldwide Events',
                fill='tozeroy',
                hovertemplate ="year: <b>%{x}</b><br>Region: World<br>Events:%{x}<br>",
             ),
            row=3, col=1)
fig.add_trace(go.Scatter(
                x=dfPopWorld['year'],
                y=dfPopWorld['pop'],
                name='World Populaiton',
                hovertemplate ="year: <b>%{x}</b><br>Region: World<br>Population:%{x}<br>",
             ),
              row=3, col=1,secondary_y=True,)

fig.update_xaxes(tickvals = yearList,
                 tickangle=20,
                 tickfont_size=10, 
                 row=3, col=1)
#------------------------------------------------------------------------------
fig.update_layout(width=1000,                  
                  height=620, 
                 )

fig.update_layout(
    xaxis_tickfont_size=10, 
    yaxis_tickfont_size=10, 
    title="Events Increase vs Populaiton Growth",
    legend=dict(
        x=1.02
    ),
    template='plotly_dark',
)
fig.update_layout(
    yaxis=dict(
        title="Events Count",
        tickfont_color ="rgb(99,110,250)",
    ),
    yaxis2=dict(
        tickfont_color="rgb(229,82,57)",
    ),
    yaxis3=dict(
        tickfont_color="rgb(0,204,150)",
    ),
    yaxis4=dict(
        title="Population",
        tickfont_color="rgb(148,86,217)",
    ),
    yaxis5=dict(
        title="Events Count",
        tickfont_color="rgb(255,161,90)",
    ),
    yaxis6=dict(
        tickfont_color="rgb(25,211,243)",
    ),
    yaxis7=dict(
        tickfont_color="rgb(255,102,146)",
    ),
    yaxis8=dict(
        title="Population",
        tickfont_color="rgb(182,232,128)",
    ),
    yaxis9=dict(
        title="Events Count",
        tickfont_color="rgb(255,151,255)",
    ),
    yaxis10=dict(
        title="Population",
        tickfont_color="rgb(254,203,82)",
    ),
)
fig.update_traces(hoverlabel_font_color='white',)
fig.show()
fig.write_html("hw2_Graph2.html")
```
