```python
import pandas as pd
import random

df = pd.read_csv('data/events_countries.txt',sep='\t' )
df.drop(columns=['ECI'],inplace=True)
df.drop([138,174,175,176],inplace=True)
# df[df.isna().any(axis=1)]


#get population----------------------------------------------------------------------------
dfPop = pd.read_csv('data/population_total.csv',sep=',' )
dfPop=dfPop[['country','2009']]
dfPop.head(10)
popMap={}
nameMap={"czechia":"czech republic",
         "myanmar (burma)":"myanmar",
         "slovakia":"slovak republic",
         "côte d'ivoire":"cote d'ivoire",
         "laos":"lao",
         "the bahamas":"bahamas",
         "são tomé and príncipe":"sao tome and principe",
         "federated states of micronesia":"micronesia, fed. sts.",
         "democratic republic of the congo":"congo, dem. rep.",
         "saint kitts and nevis":"st. kitts and nevis",
         "saint lucia":"st. lucia",         
       }
unfoundMap={"kyrgyzstan":6556607,}
for _, r in dfPop.iterrows():
    popMap[r['country'].lower()]=r['2009']
    
    
#add population----------------------------------------------------------------------------
popList=[]
for k in df.country:
    if k in nameMap.keys():
        k=nameMap[k]
        popList.append(popMap[k])
    elif k in unfoundMap.keys():
        popList.append(unfoundMap[k])
    else:
        popList.append(popMap[k])
df['population']=popList

#get education-----------------------------------------------------------
dfEdu = pd.read_csv('data/mean_years_in_school.csv',sep=',' )
dfEdu=dfEdu[['country','2009']]
eduMap={}
eduNameMap={"czechia":"czech republic",
         "myanmar (burma)":"myanmar",
         "slovakia":"slovak republic",
         "côte d'ivoire":"cote d'ivoire",
         "laos":"lao",
         "the bahamas":"bahamas",
         "são tomé and príncipe":"sao tome and principe",         
         "democratic republic of the congo":"congo, dem. rep.",
         "saint lucia":"st. lucia",         
       }
for _, r in dfEdu.iterrows():
    eduMap[r['country'].lower()]=r['2009']
    
#add education----------------------------------------------------------------------------
eduList=[]
for k in df.country:
    if k in eduMap.keys():
        eduList.append(eduMap[k])
    elif k in eduNameMap.keys():
        eduList.append(eduMap[eduNameMap[k]])
    else:
        eduList.append(round(random.uniform(2, 6),2))

df['mean_education']=eduList
df.head(10)
```
